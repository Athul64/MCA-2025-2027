## **2. Display future leap years from current year to a final year entered by user.**

````python
current_year = int(input("Enter the current year: "))
future_year = int(input("Enter the future year: "))

print("\nleap years from",current_year,"to",future_year,"are ")
for year in range(current_year, future_year + 1):
  if(year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year)
````

## **4. Count the occurrences of each word in a line of text.**


````python
text = input("Enter a line of text: ").split()
occu = {}
for ch in text:
  if ch in occu:
    occu[ch] += 1
  else:
    occu[ch] = 1
print("Word occurrences:")
for text, count in occu.items():
    print(f"{text}: {count}")
````

## **7. Enter 2 lists of integers. Check (a) Whether list are of same length (b) whether list sums to same value (c) whether any value occur in both**


````python
list_a = list(map(int, input("Enter first list of integers separated by space : ").split()))
list_b = list(map(int, input("Enter second list of integers separated by space : ").split()))

# Check if lists are of same length
same_len = len(list_a) == len(list_b)

# Check if lists sum to same value
same_sum = sum(list_a) == sum(list_b)

# Check if any value occurs in both
common_val = set(list_a) & set(list_b)

# Display results
print(f"\nSame length: {same_len}")
print(f"Same sum: {same_sum}")
print(f"Common values: {list(common_val) if common_val else 'None'}")
````

## **19. Find gcd of 2 numbers.**


````python
gcd_1 = int(input("Enter a number: "))
gcd_2 = int(input("Enter a number: "))

if gcd_1 > gcd_2:
  smaller = gcd_2
else:
  smaller = gcd_1

for i in range(1, smaller + 1):
  if(gcd_1 % i == 0) and (gcd_2 % i == 0):
    gcd = i
print("The GCD of", gcd_1, "and", gcd_2, "is:", gcd)
````

## **4. Generate a list of four digit numbers in a given range with all their digits even and the number is a perfect square.**


````python
even_square = []
for num in range(32, 100):
  square = num * num
  if 1000 <= square <= 9999:
    digit = str(square)
    if all(int(d) % 2 == 0 for d in digit):
      even_square.append(square)
print("Four digit perfect square with all even number: ",even_square)
````

````python
class Rectangle:
    def __init__(self, l, b):
        self.l = l
        self.b = b

    def area(self):
        return self.l * self.b

    def perimeter(self):
        return 2 * (self.l + self.b)


# Input
l1, b1 = map(int, input("Enter length and breadth of first rectangle: ").split())
l2, b2 = map(int, input("Enter length and breadth of second rectangle: ").split())

# Create objects
rect_1 = Rectangle(l1, b1)
rect_2 = Rectangle(l2, b2)

# Output
print("Area of rectangle 1:", rect_1.area())
print("Perimeter of rectangle 1:", rect_1.perimeter())

print("Area of rectangle 2:", rect_2.area())
print("Perimeter of rectangle 2:", rect_2.perimeter())

print("\nCompare are of two rectangle")

if rect_1.area() == rect_2.area():
  print("Value are Same..")
elif rect_1.area() > rect_2.area():
  print("Rectangle 1 is greater ",rect_1.area())
else:
    print("Rectangle 2 is greater ",rect_2.area())
````

## **2. Python program to copy odd lines of one file to other**

````python
with open("text.txt", "r") as src, open("dest.txt", "w") as des:
    for lno, line in enumerate(src, start=1):
        if lno % 2 != 0:   # odd lines
            try:
                des.write(line)
            except:
                print("Something went wrong")

print("Successfully")
````
